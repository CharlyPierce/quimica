```python
from PIL import Image
import matplotlib.pyplot as plt

# Leer la imagen '0.png'
imagen = Image.open('0.png')

# Mostrar la imagen en el notebook con matplotlib
plt.imshow(imagen)
plt.axis('off')  # Oculta los ejes
plt.show()
```
